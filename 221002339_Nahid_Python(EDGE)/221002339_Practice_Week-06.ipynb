```python
#1

class Book:
    def __init__(self, title, author, year):
        self.title = title
        self.author = author
        self.year = year

    def get_details(self):
        print(f"Title: {self.title}, Author: {self.author}, Year Published: {self.year}")

book = Book("Into The Wild", "Nahid", 2010)
book.get_details()
```

```python
#2

class Animal:
    def __init__(self, name, sound):
        self.__name = name
        self.__sound = sound

    def get_name(self):
        return self.__name

    def set_name(self, name):
        self.__name = name

    def get_sound(self):
        return self.__sound

    def set_sound(self, sound):
        self.__sound = sound

    def speak(self):
        print(f"{self.__name} says {self.__sound}")

animal = Animal("Dog", "Woof")
animal.speak()
animal.set_sound("Bark")
animal.speak()
```

```python
#3

class BankAccount:
    def __init__(self, initial_balance=0):
        self._balance = initial_balance

    def deposit(self, amount):
        if amount > 0:
            self._balance += amount
            print(f"Deposited {amount}. New balance is {self._balance}.")
        else:
            print("Deposit amount must be positive.")

    def withdraw(self, amount):
        if amount <= 0:
            print("Withdrawal amount must be positive.")
        elif amount > self._balance:
            print("Insufficient funds for this withdrawal.")
        else:
            self._balance -= amount
            print(f"Withdrew {amount}. New balance is {self._balance}.")

    def get_balance(self):
        return self._balance
account = BankAccount(100)
account.deposit(50)
account.withdraw(120)
account.withdraw(80)
print("Final balance:", account.get_balance())
```

```python
#4

import math

class Shape:
    def area(self):
        raise NotImplementedError("The area method must be implemented by subclasses.")

class Circle(Shape):
    def __init__(self, radius):
        self.radius = radius

    def area(self):
        return math.pi * (self.radius ** 2)

class Rectangle(Shape):
    def __init__(self, length, width):
        self.length = length
        self.width = width

    def area(self):
        return self.length * self.width

shapes = [Circle(5), Rectangle(10, 4)]

for shape in shapes:
    print(f"The area of the {shape.__class__.__name__} is: {shape.area():.2f}")

```

```python
#5

class MathOperations:

    @staticmethod
    def multiply(x, y):
        return x * y

result = MathOperations.multiply(10, 10)
print(result)
```

```python
#6

class Calculator:
    def add(self, *args):
        if len(args) == 1 and isinstance(args[0], list):
            return sum(args[0])
        elif len(args) == 2:
            return args[0] + args[1]
        else:
            raise ValueError("Invalid! Enter two numbers or a list of numbers.")

calc = Calculator()
result1 = calc.add(10, 20)
print("Sum of 10 and 20:", result1)
result2 = calc.add([1, 2, 3, 4, 5])
print("Sum of list [1, 2, 3, 4, 5]:", result2)
```

```python
#8

class Car:
    def __init__(self, make, model, year):
        self.make = make
        self.model = model
        self.year = year

    def __repr__(self):
        return f"Car(make={self.make}, model={self.model}, year={self.year})"

car1 = Car("Toyota", "Corolla", 2020)
car2 = Car("Tesla", "Model S", 2023)

print(car1)
print(car2)

```

```python
#9

class Vehicle:
    def __init__(self, make, model, year):
        self.make = make
        self.model = model
        self.year = year

    def print_details(self):
        print(f"Make: {self.make}")
        print(f"Model: {self.model}")
        print(f"Year: {self.year}")

class Truck(Vehicle):
    def __init__(self, make, model, year, payload_capacity):
        super().__init__(make, model, year)
        self.payload_capacity = payload_capacity

    def print_details(self):
        super().print_details()
        print(f"Payload Capacity: {self.payload_capacity} kg")

truck = Truck("Ford", "F-150", 2023, 1000)
truck.print_details()
```
