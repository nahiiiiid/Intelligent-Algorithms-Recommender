```python
#3

import heapq

nums = [4, 1, 7, 3]
heapq.heapify(nums)
heapq.heappush(nums, 2)
print(heapq.heappop(nums))
```

```python
#4

from collections import deque

dq = deque([1, 2, 3])
dq.append(4)
dq.appendleft(0)
dq.pop()
dq.popleft()
```

```python
# Stack implementation

class stack:
    def __init__(self):
        self.stack = []

    def push(self, x):
        self.stack.append(x)

    def pop(self):
        return self.stack.pop()

    def top(self):
        return self.stack[-1]

myStack = stack()
myStack.push(10)
myStack.push(20)

print(myStack.pop())
print(myStack.top())
```
