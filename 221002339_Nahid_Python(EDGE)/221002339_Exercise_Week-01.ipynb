```python
# solution of problem-02

a = int(input("Enter first number: "))
b = int(input("Enter second number: "))

print("Sum:", a + b)
print("Difference (a - b):", a - b)
print("Difference (b - a):", b - a)
print("Product:", a * b)
print("Quotient:", a / b)
```

```python
# solution of problrm-03

length = float(input("Enter the length of a rectangle: "))
width = float(input("Enter the width of a rectangle: "))

area = (length * width)

print("The area of the rectangle is ", area)
```

```python
# solution of problem-08

numbers = [10, 200, 30, 400, 500]

minimum_number = min(numbers)
maximum_number = max(numbers)
sumOfNumbers = sum(numbers)

print("The minimum number of the list = ", minimum_number)
print("The maximum number of the list = ", maximum_number)
print("The sum of the numbers of the list = ", sumOfNumbers)
```
