```python
#9

stack = []

def push():
    number = int(input("Enter a number to push: "))
    stack.append(number)
    print(f"The number {number} pushed in the stack!")

def pop():
    if stack:
        number = stack.pop()
        print(f"The number {number} popped from the stack!")
    else:
        print("The stack in empty and can't pop!")

def display():
    print("The current stack: ", stack)

while True:
    print("\nChoose an option (push, pop, display, quit)")
    user_choice = input("Enter what do you want? ")

    if user_choice == "push":
        push()
    elif user_choice == "pop":
        pop()
    elif user_choice == "display":
        display()
    elif user_choice == "quit":
        break;
    else:
        print("You choose invalid option!")
```

```python
#10

import random

random_numbers = [random.randint(1, 100) for _ in range(10)]
print("Random Numbers list: ", random_numbers)

max_num = max(random_numbers)
min_num = min(random_numbers)
avg_num = sum(random_numbers) / len(random_numbers)

print("The maximum number: ", max_num)
print("The minimum number: ", min_num)
print("The average number: ", avg_num)
```
