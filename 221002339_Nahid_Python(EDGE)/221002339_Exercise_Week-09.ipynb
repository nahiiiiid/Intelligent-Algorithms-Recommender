```python
#3 - Indexing

array = np.array([10, 20, 30, 40])
print(array[1])

array2d = np.array([[1, 2], [3, 4]])
print(array2d[1, 1])
```

```python
#3 - Slicing

array = np.array([10, 20, 30, 40, 50])
print(array[1:4])

array2d = np.array([[1, 2, 3], [4, 5, 6]])
print(array2d[:, 1])
```

```python
#3 - Boolean Indexing

array = np.array([10, 20, 30, 40])
print(array[array > 20])
```

```python
#4 - Reshaping

array = np.arange(6)
reshaped = array.reshape(2, 3)
print(reshaped)
```

```python
#4 - Stacking

a = np.array([1, 2])
b = np.array([3, 4])

vstacked = np.vstack((a, b))
print(vstacked)

hstacked = np.hstack((a, b))
print(hstacked)
```

```python
#4 - Splitting

array = np.arange(6).reshape(2, 3)
print(np.hsplit(array, 3))
```
