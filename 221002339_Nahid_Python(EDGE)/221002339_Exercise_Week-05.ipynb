```python
#1

def find_largest(a, b, c):
    return max(a, b, c)

print(find_largest(5, 10, 3))
```

```python
#2

def factorial(n):
    if n == 0:
        return 1
    else:
        return n * factorial(n-1)

print(factorial(5))
```

```python
#4

class Rectangle:
    def __init__ (self, length, width):
        self.length = length
        self.width = width

    def area(self):
        return self.length * self.width

    def perimeter(self):
        return 2 * (self.length + self.width)

rect = Rectangle(10, 5)
print(rect.area())
print(rect.perimeter())
```

```python
#5

class BankAccount:
    def __init__ (self, account_number, balance=0):
        self.account_number = account_number
        self.balance = balance

    def deposit(self, amount):
        self.balance += amount

    def withdraw(self, amount):
        if self.balance >= amount:
            self.balance -= amount
        else:
            return "Insufficient funds"

    def get_balance(self):
        return self.balance

account = BankAccount(101, 1000)
account.deposit(500)
print(account.get_balance())
account.withdraw(200)
print(account.get_balance())
```

```python
#6

class Bank:
    def __init__ (self):
        self.accounts = []

    def add_account(self, account):
        self.accounts.append(account)

    def total_funds(self):
        return sum(account.get_balance() for account in self.accounts)

account1 = BankAccount(101, 1000)
account2 = BankAccount(102, 1500)

bank = Bank()
bank.add_account(account1)
bank.add_account(account2)

print(bank.total_funds())
```

```python
#7

def sum_of_evens(numbers):
    return sum(n for n in numbers if n % 2 == 0)

numbers = list(map(int, input("Enter numbers: ").split()))
print(sum_of_evens(numbers))
```

```python
#8

def is_prime(n):
    if n < 2:
        return False
    for i in range(2, int(n**0.5) + 1):
        if n % i == 0:
            return False
    return True

print(is_prime(11))
print(is_prime(15))
```

```python
#9

class Student:
    def __init__ (self, name, age, grades):
        self.name = name
        self.age = age
        self.grades = grades

    def display(self):
        print(f"Name: {self.name}, Age: {self.age}")

    def average_grade(self):
        return sum(self.grades) / len(self.grades)

student = Student("Alice", 20, [85, 90, 78])
student.display()
print(student.average_grade())
```

```python
#10

def get_common_elements(list1, list2):
    return list(set(list1).intersection(list2))

print(get_common_elements([1, 2, 3, 4], [3, 4, 5, 6]))
```

```python
#11

class Vehicle:
    def __init__(self, brand, speed=0):
        self.brand = brand
        self.speed = speed

    def accelerate(self, amount):
        self.speed += amount
        print(f"Speed increased to {self.speed}")

    def brake(self, amount):
        self.speed = max(0, self.speed - amount)
        print(f"Speed decreased to {self.speed}")

car = Vehicle("Toyota", 50)
car.accelerate(20)
car.brake(30)
```

```python
#12

def get_unique_elements(lst):
    return list(set(lst))

print(get_unique_elements([1, 2, 2, 3, 4, 4, 5]))
```

```python
#13

class Library:
    def __init__(self):
        self.books = []

    def add_book(self, book):
        self.books.append(book)
        print(f"'{book}' added to the library.")

    def remove_book(self, book):
        if book in self.books:
            self.books.remove(book)
            print(f"'{book}' removed from the library.")
        else:
            print(f"'{book}' not found in the library.")

    def display_books(self):
        print("books in library: ", self.books)

library = Library()
library.add_book("The Great Gatsby")
library.add_book("1984")
library.display_books()
library.remove_book("1984")
library.display_books()
```
