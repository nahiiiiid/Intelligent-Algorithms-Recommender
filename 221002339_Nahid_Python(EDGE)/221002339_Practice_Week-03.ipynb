```python
#1

number = int (input("Enter an integer number: "))

if (number % 3 == 0) and (number % 5 == 0):
    print("Divisible by 3 and 5")
else:
    print("Not divisible by 3 and 5")
```

```python
#2

def sumOfDigits(n):
    return sum(int(digit) for digit in str(abs(n)))

number = int(input("Enter an integer number: "))
print(f"The sum of all digits of the number: {sumOfDigits(number)}")
```

```python
#3
even_count = 0
odd_count = 0

while True:
    number = int(input("Enter numbers: "))

    if number == 0:
        break

    if number % 2 == 0:
        even_count += 1
    else:
        odd_count +=1
print(f"Total even number: {even_count}")
print(f"Total odd number: {odd_count}")
```

```python
#4

fact = 1

number = int (input("Enter a number to calculate factorial: "))

for i in range(1, number+1):
    fact = fact * i
print(f"The factorial of {number} = {fact}")
```

```python
#5


number = int(input("Enter a number for multiplication table: "))

if number > 12:
    print("The number you enetered is greater than 12!")

else:

    print(f"Multiplication table for {number}: ")

    for i in range(1, 11):
        print(f"{number} x {i} = {number * i}")
```

```python
#6

numbers = []
while True:
    number = int (input("Enter numbers: "))

    if number == -1:
        break
    else:
        numbers.append(number)

if numbers:
    largest = max(numbers)
    print(f"The largest number is: {largest}")
else:
    print("The list is empty!")
```

```python
#7

number = int(input("Enter a nunber: "))
is_prime = True
if number > 1:
    for i in range(2, int(number ** 0.5) + 1):
        if number % i == 0:
            is_prime = False
            break
    if is_prime:
        print("Prime")
    else:
        print(f"Not Prime")
else:
    print(f"Not Prime")
```

```python
#8

grade = int (input("Enter grade: "))

if (grade >= 90) and (grade <= 100):
    print("A")
elif (grade >= 80) and (grade <= 89):
    print("B")
elif (grade >= 70) and (grade <= 79):
    print("C")
elif (grade >= 60) and (grade <= 69):
    print("D")
elif (grade < 60):
    print("F")
```

```python
#9

str = input("Enter a string: ")

vowel_count = 0

vowels = "aeiouAEIOU"

for char in str:
    if char in vowels:
        vowel_count += 1

print(f"Total vowel present in the string {str} = {vowel_count}")
```

```python
#10

balance = 0

while True:
    print("\n1. Check Balance")
    print("2. Deposit")
    print("3. Withdraw")
    print("4. Exit")

    choice = input("Choose an option: ")

    if choice == "1":
        print(f"Your balance is: {balance}")

    elif choice == "2":
        amount = float(input("Amount to deposit: "))
        balance += amount
        print(f"Deposited: {amount}")

    elif choice == "3":
        amount = float(input("Amount to withdraw: "))
        if amount <= balance:
            balance -= amount
            print(f"Withdraw: {amount}")
        else:
            print("Insufficient balance.")
    elif choice == "4":
        break;

```
